```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#reading data
```

```python
data=pd.read_csv("High_School_Public_Results_2022_EG_first_attempt.csv")
data
```

```python
#analyzing data
```

```python
data.info()
```

```python
data[['Percentage']].describe()
```

```python
# here we get that (1) About 75% of the students are above 50% percentage grade
#                  (2) Most of the students passed 
#                  (3) Only 25% is above 73% percentage grade
#                  (4) Max student with a percentage grade = 99.27%
```

```python
# 'school_name','administration','city', 'branch', 'status','gender'
# we need to convert this columns to categories to get further insights 
```

```python
data['school_name']=pd.Categorical(data.school_name)
data['administration']=pd.Categorical(data.administration)
data['city']=pd.Categorical(data.city)
data['branch']=pd.Categorical(data.branch)
data['status']=pd.Categorical(data.status)
data['gender']=pd.Categorical(data.gender)
```

```python
data.info()
```

```python
data.describe(include="category")
```

```python
#here we understand that (1) We had a total of 682348 final year students last year
#                        (2) School with the most students is "السعيدية الثانوية العسكرية"
#                        (3) Top administration having students is "المنتزه"
#                        (4) City with most student is "Cairo"
#                        (5) Most of the students are "علمي علوم"
#                        (6) Most of the students are Females
```

```python
data["city"].value_counts()
```

```python
data.rename(columns = {'city':'Government'}, inplace = True)
data.rename(columns = {'administration':'City'}, inplace = True)
data
```

```python
plt.figure(figsize=(10,15))
keys = ["Cairo", "Giza", "Sharqiya","Daqahliya" , "Alexandria" , "Qalioubiya", "Minya", "Gharbiya", "Monofiya", "Beheira",
        "Assiut", "Kafr El Sheikh", "Sohag", "Bani Suef" ,"Fayyoum", "Qena", "Damietta", "Ismailia", "Aswan", "Luxor"]
plt.pie(data["Government"].value_counts().iloc[:20] , labels = keys , autopct='%1.0f%%')
plt.title("Top 20 Cities Distribution of Thanawya Amma Students in Egypt")
plt.show()
```
